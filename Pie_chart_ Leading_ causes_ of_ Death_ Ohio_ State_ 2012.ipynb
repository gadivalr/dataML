```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Data from Ohio State, Leading casuses of death in the state from year 2012
cause='Chronic Diseases','Unitentional Injuries','Alzheimers','Influenza and Pneumonia','Sepsis','Others'
#percentile
percentile=[62,5,4,2,1,26]
```

```python
#set the pie chart plot
plt.figure(figsize=(7,7))
#explode the largest pie in the data set
explode=(0.05,0,0,0,0,0)
#set pie chart properties 
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.1f%%', startangle=70)
plt.axis('equal')
plt.title('Ohio State-2012: Leading causes of Death')

plt.show()
```
