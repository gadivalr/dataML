```python
import random as rd
dado=[1,2,3,4,5,6]

for i in range(5):
    M=[]
    N=[]
    six=0
    four=0
    otro=0
    for i in range(10000):
        rand= rd.choice(dado)
        if rand==6:
            six +=1
        else:
            otro+=1
    for i in range(1000):
        rand=rd.choice(dado)
        if rand==4:
            four+=1
    M.append(six)
    N.append(four)
M1=int(M[0])
N2=int(N[0])
print(M1/10000,N2/1000)
```
